# 25. Data Visualization with Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic.plot()
plt.show()
```

```python
titanic.plot(subplots = True, figsize= (12,8), sharex = False, sharey = False) #have own x , sharey default false
plt.show()
```

```python
titanic.age.plot(figsize= (12,8))
plt.show()
```
